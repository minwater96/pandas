### to_numeric

```python
# to_numeric => (df[], errers) 수치형 변수로 바꿔주며, errors를 통해 중간에 글이 있는 값을 NaN으로 바꿔줌
pd.to_numeric(df1['운동량'], errors='coerce')
```

```python
df1['운동량'] = pd.to_numeric(df1['운동량'], errors='coerce')
```

```python
df1.head()
```

```python
bins = [0, 1000, 2000, df1['운동량'].max()]

pd.cut(df1['운동량'], bins).value_counts() # 어떤 data를 기준에 따라 분류
```

```python
pd.cut(df1['운동량'], bins, labels=['부족', '보통', '많음'])
```

```python
df1['운동량_cut'] = pd.cut(df1['운동량'], bins=10)
df1['운동량_cut'].value_counts()
```

```python
# qcut => 일정한 비율을 통해 나눠줌
# df1['운동량_qcut'] = pd.qcut(df1['운동량'], q=10)
df1['운동량_qcut'].value_counts()
```

```python
# 0~20%, 20%~80%, 80%~100%
qcut_bins = [0, 0.2, 0.8, 1]
pd.qcut(df1['운동량'], qcut_bins)
```

```python
pd.qcut(df1['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
```
