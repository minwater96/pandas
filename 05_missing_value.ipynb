### fillna

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
# 결측치에 해당 값 대체
df1['age'].fillna(100)
```

```python
df1['age'] = df1['age'].fillna(100)
```

```python
df1.tail()
```

```python
df1['deck'].fillna('C')
```

```python
# categories type은 categorie에 없는 정보는 넣을 수 없음
# df1['deck'].fillna('No Data')
```

```python
df1['deck'] = df1['deck'].cat.add_categories('No Data')
df1['deck'].fillna('No Data')
```

```python
df1 = df.copy()
```

```python
df1['age'] = df1['age'].fillna(df['age'].mean())
df1['age'].tail()
```

```python
df1['age'] = df1['age'].fillna(df['age'].median())
df1['age'].tail()
```

- 남자 승객의 결측치 => 남자 승객의 평균
- 여자 승객의 결측치 => 여자 승객의 평균

```python
cond1 = df1['sex'] == 'male' 
df_M = df1[cond1]
print(df_M['age'].mean())
df_M['age'] = df_M['age'].fillna(df_M['age'].mean())
df_M.head()
```

```python
cond2 = df1['sex'] == 'female' 
df_F = df1[cond2]
#print(df_F['age'].mean())
df_F['age'] = df_F['age'].fillna(df_F['age'].mean())
df_F.tail()
```

```python
df2 = pd.concat([df_M, df_F])
print(round(df2['age'].mean(), 3))
```
