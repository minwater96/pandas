## concat 

```python
import pandas as pd
```

```python
sales_df1 = pd.read_csv('data/sales_data1.csv')
sales_df2 = pd.read_csv('data/sales_data2.csv')
```

```python
sales_df1.head()
```

```python
sales_df2.head()
```

```python
sales_df = pd.concat([sales_df1, sales_df2], ignore_index=True)
```

```python
sales_df.iloc[170:190]
```

```python
sales1 = sales_df1.iloc[:, :5]
sales2 = sales_df2.iloc[:, :5]
```

```python
sales1.head()
```

```python
pd.concat([sales1, sales2], axis=1).head()
```

### merge

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}
```

```python
money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
info_df = pd.DataFrame(info)
money_df = pd.DataFrame(money)
```

```python
print(info_df.head())
print(money_df.head())
```

```python
pd.merge(info_df, money_df)
```

```python
pd.merge(info_df, money_df, how='outer')
```

```python
pd.merge(info_df, money_df, how='left')
```

```python
pd.merge(info_df, money_df, how='right')
```

```python
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}
money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
info_df = pd.DataFrame(info)
money_df = pd.DataFrame(money)
```

```python
pd.merge(info_df, money_df, left_on='이름', right_on='고객명')
```
